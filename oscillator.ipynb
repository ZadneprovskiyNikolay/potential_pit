Собственные зачения и собственные функции для осциляторного потенциала

```python
import numpy as np
from scipy.linalg import eigh
import matplotlib.pyplot as plt
```

```python
m = 9.109e-31
omega = 1
a = 1
x_min = -a
x_max = a
```

Аппроксимация оператора Гамильтона:

<tex>$$\hat H = - {{\hbar}\over{2m}}{{d^2}\over{dx^2}} + {{1}\over{2}}m \omega^2x^2$$<tex>

Заменить производных второго порядка разностными отношениями($h$ - шаг сетки):

<tex>$${{d^2}\over{dx^2}} \approx {{f(x + h) - 2f(x) + f(x - h)}\over{h^2}}$$<tex>

```python
N = 1000 # количество точек для вычисления собственных функций
h = a / N # шаг сетки

# Матрица оператора Гамильтона
H = np.zeros((N, N))
for i in range(N):
    H[i, i] = 1 / h**2 + 0.5* m*(omega**2) * ((i * h)**2)
    if i > 0:
        H[i, i - 1] = -0.5 / (h**2)
        H[i - 1, i] = -0.5 / (h**2)

# Вычисление собственных значений и собственных векторов оператора Гамильтона
eigenvalues, eigenvectors = eigh(H)

x = np.linspace(-a, a, N)
for i in range(5):
    plt.plot(x, eigenvectors[:, i], label='n={}'.format(i))
    print(f'E{i+1}: {eigenvalues[i]}')
plt.legend()
plt.xlabel('x')
plt.ylabel('Psi')
plt.show()
```
